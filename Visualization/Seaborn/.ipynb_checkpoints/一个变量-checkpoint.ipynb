```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as sci
```

# 分布

## 直方图

```python
tips = sns.load_dataset('tips')

sns.distplot(tips['tip'], kde=True);
```

## 毛毯图

```python
sns.distplot(tips['total_bill'], bins=30, kde=False, rug=True);
```

## 多个子图

```python
fig, axes = plt.subplots(1, 2, figsize=(10,5))

sns.distplot(tips['total_bill'], ax=axes[0], kde=False)
sns.rugplot(tips['total_bill'], ax=axes[0])

sns.kdeplot(tips['total_bill'], ax=axes[1], shade=True)
sns.rugplot(tips['total_bill'], ax=axes[1])
```

## 密度分布图

```python
# 单个样本数据密度分布图
plt.figure(figsize=(8,4))

sns.kdeplot(tips['total_bill'], label='auto')
sns.kdeplot(tips['total_bill'], bw=10, label="bw: 10",linewidth = 1.5)
sns.kdeplot(tips['total_bill'], bw=100, label="bw: 100",linestyle = '--',linewidth = 1.5)
# bw → 确定核的大小，近似理解为拟合程度，bw越大，曲线越平缓

plt.show()
```
