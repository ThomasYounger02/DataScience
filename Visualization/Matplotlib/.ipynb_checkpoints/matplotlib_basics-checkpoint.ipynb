# 基础绘图

```python
import numpy as np
import matplotlib.pyplot as plt

# 中文处理
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

## 绘图属性

```python
# Data
x=[1, 2, 3, 4]
y=[1, 4, 9, 16]

# Plot
plt.plot(x, y, color='r', marker='o', linestyle='dashed', label='legend')  # 红色线条，圆圈标记，虚线连接

#  Decoration
plt.axis([0, 6, 0, 20])                                   # [x轴最小值, x轴最大值, y轴最小值, y轴最大值]
plt.xlabel('x_axis_label')                                # x轴标签
plt.ylabel('y_axis_label')                                # y轴标签
plt.title('Title')                                        # 标签
plt.legend()                                              # 图例

plt.annotate('中文注释', xy=(2,5), xytext=(2, 10),              # xy：箭头注释中箭头所在位置，xytext：注释文本所在位置
            arrowprops=dict(facecolor='black', shrink=0.01)     # arrowprops在xy和xytext之间绘制箭头, shrink表示注释点与注释文本之间的图标距离
            )                                              
# Show
plt.show()
```

## 多图绘制

```python
# Data
x = np.linspace(0, 10, 30)

# Plot
fig = plt.figure()                        # 创建1个画板

ax1 = plt.subplot(2,1,1)                  # 创建画纸，并选择画纸1
plt.plot(x, np.sin(x), 'r', 'sin(x)')         # 在画纸1上作图


ax2 = plt.subplot(2,1,2)                 # 选择画纸2
plt.plot(x, np.cos(x), 'k', 'cos(x)')        # 在画纸2上作图

# Show
plt.show()
```

# 绘制一张图

```python
# Data
x = np.linspace(0, 10, 30)

# Plot
plt.plot(x, np.sin(x), '--', label='sin')    # 虚线
plt.plot(x, np.cos(x), '-', label='cos')    # 实线

# Decoration
plt.xlim(0, 10)                         # x轴范围
plt.ylim(-1.5, 1.5)                     # y轴范围
plt.xlabel('x_lable')                   # x轴标签
plt.ylabel('y_lable')                   # y轴标签
plt.title('Title')                      # 标题
plt.grid()                              # 显示网格
plt.legend(loc='lower left', frameon=True, ncol=2)    # 左下角，显示边框，分成两列

plt.axhline(y=0.8, ls='--', c='r')                               # 绘制y=0.8的水平参考线
plt.axvspan(xmin=4, xmax=6, facecolor='r', alpha=0.3)            # 垂直x轴，参考区域
plt.axhspan(ymin=-0.2, ymax=0.2, facecolor='y', alpha=0.3);      # 垂直y轴，参考区域
plt.text(3.2, 0, 'sin(x)', weight='bold', color='r')              # 注释文字
plt.annotate('maximum', xy=(np.pi/2, 1), xytext=(np.pi/2+1, 1),
             weight='bold',
             color='r',
             arrowprops=dict(arrowstyle='->', connectionstyle='arc3', color='r'))   # 箭头标识

# Show
plt.show()
```

```python
# Data
t = np.arange(0, 5, 0.2)

# 线条1
x1=y1=t
# 线条2
x2=x1
y2=t**2
# 线条3
x3=x1
y3=t**3

# 使用plot绘制线条
linesList=plt.plot(
    x1, y1,  
    x2, y2,  
    x3, y3 )

# 用setp方法可以同时设置多个线条的属性
plt.setp(linesList, color='r')
plt.show()
```

# 绘制多个子图

```python
fig, ax_lst = plt.subplots(2, 2)        # 创建一个以 axes为单位的 2x2 网格的 figure 
plt.show()
```

```python
# 在2个子图中，显示sin(x)和cos(x)的图像
fig = plt.figure()
ax1 = fig.add_axes([0.1, 0.5, 0.8, 0.4], ylim=(-1.2, 1.2))
ax2 = fig.add_axes([0.1, 0.1, 0.8, 0.4], ylim=(-1.2, 1.2))

x = np.linspace(0, 10)
ax1.plot(np.sin(x));
ax2.plot(np.cos(x));
```

```python
# 用for创建6个子图，并且在图中标识出对应的子图坐标
for i in range(1, 7):
    plt.subplot(2, 3, i)
    plt.text(0.5, 0.5, str((2, 3, i)), fontsize=18, ha='center')
```
